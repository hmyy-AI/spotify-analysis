```python
df_tracks["duration"]= df_tracks["duration_ms"].apply (lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_tracks.duration.head()
```

```python
# 1. Tính toán tương quan (giữ nguyên vì dòng này bạn đã làm đúng)
corr_df = df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson", numeric_only=True)

# 2. Thiết lập khung hình
plt.figure(figsize=(14, 6))

# 3. Vẽ heatmap - Sửa fmt=".ig" thành fmt=".1f"
heatmap = sns.heatmap(corr_df, annot=True, fmt=".4f", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")

# 4. Đặt tiêu đề
heatmap.set_title("Correlation Heatmap between Variables")

# 5. Xoay nhãn trục X (Sửa lại cú pháp cho đúng)
plt.xticks(rotation=90)

# 6. Hiển thị
plt.show()
```

```python
#Insight: Mối tương quan mạnh nhất là giữa Energy (Năng lượng) và Loudness (Độ ồn) với chỉ số xấp xỉ 0.76. 
#Ngược lại, Acousticness (Độ mộc) có tương quan âm rất mạnh với Energy (-0.71).
# Điều này có nghĩa là bài hát càng "mộc" thì năng lượng càng thấp và ngược lại.
```

```python
# 1. Lấy giá trị từ index
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')

# 2. Sửa lỗi bằng cách thêm format='mixed'
# Lệnh này bảo Pandas: "Dòng nào có năm thì lấy năm, dòng nào có đủ ngày thì lấy đủ"
df_tracks['dates'] = pd.to_datetime(df_tracks['dates'], format='mixed')

# 3. Lấy năm
years = df_tracks['dates'].dt.year

# Kiểm tra thử
df_tracks[['dates']].head()
```

```python
# 2. Vẽ biểu đồ phân bố số lượng bài hát theo năm
import seaborn as sns
import matplotlib.pyplot as plt

# Thiết lập kiểu biểu đồ
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")

# Hiển thị biểu đồ
plt.show()
#Insight: Số lượng bài hát tăng trưởng ổn định từ năm 1920,
# nhưng bắt đầu "nhảy vọt" kể từ sau năm 1990 và đạt đỉnh vào những năm 2020 với hơn 12.000 bài mỗi năm.
# Điều này phản ánh sự phát triển của công nghệ thu âm và các nền tảng streaming.
```

```python
# 1. Lấy dữ liệu thời lượng (cột duration bạn đã tính ở bước trước)
total_dr = df_tracks.duration

# 2. Thiết lập kích thước khung hình (rộng 18, cao 7)
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)

# 3. Vẽ biểu đồ thanh (Barplot)
# Lưu ý: errwidth nên để là số (ví dụ: 0) hoặc bỏ đi nếu dùng bản Seaborn mới
fig = sns.barplot(x=years, y=total_dr, ax=ax, errorbar=None,hue=years, palette='viridis')
fig.set(title="Year vs Duration")

# 4. Xoay nhãn năm 90 độ để không bị đè chữ
plt.xticks(rotation=90)

# 5. Hiển thị
plt.show()
#Insight: Thời lượng bài hát có xu hướng tăng dần từ năm 1920 và đạt mức cao nhất (ổn định khoảng 240-250 giây) trong giai đoạn 1990-2010.
# Tuy nhiên, từ sau năm 2010 đến nay, thời lượng bài hát đang có xu hướng giảm mạnh xuống dưới 200 giây.
```

```python
# 1. Thiết lập dữ liệu và phong cách biểu đồ
total_dr = df_tracks.duration
sns.set_style(style="whitegrid") # Thêm lưới nền trắng cho dễ nhìn

# 2. Tạo khung hình
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize=fig_dims)

# 3. Vẽ biểu đồ đường (Lineplot)
fig = sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year vs Duration")

# 4. Xoay nhãn trục X (năm) 60 độ để không bị chồng chữ
plt.xticks(rotation=60)

# 5. Hiển thị
plt.show()
#Insight: Thời lượng bài hát có xu hướng tăng dần từ năm 1920 và đạt mức cao nhất (ổn định khoảng 240-250 giây) trong giai đoạn 1990-2010.
# Tuy nhiên, từ sau năm 2010 đến nay, thời lượng bài hát đang có xu hướng giảm mạnh xuống dưới 200 giây.
```

```python
df_genre= pd.read_csv('D:/spotifydataset/SpotifyFeatures.csv')
```

```python
df_genre.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
plt.show()
#Insight: World, Classical (Cổ điển) và Jazz là những thể loại có thời lượng trung bình dài nhất (trên 250.000ms). 
#Trong khi đó, Children's Music (Nhạc thiếu nhi) có thời lượng ngắn nhất.
```

```python
# 1. Thiết lập phong cách nền tối có lưới
sns.set_style(style="darkgrid")

# 2. Tạo khung hình
plt.figure(figsize=(10, 5))

# 3. Lấy dữ liệu Top 10 (trong tiêu đề bạn ghi Top 5 nhưng code lấy head(10), bạn có thể chỉnh lại tùy ý)
famous = df_genre.sort_values("popularity", ascending=False).head(10)

# 4. Vẽ biểu đồ với tham số màu sắc (palette)
# palette='coolwarm' sẽ tạo dải màu từ xanh sang đỏ rất đẹp
sns.barplot(y='genre', x='popularity', data=famous, palette='coolwarm').set(title="Top 5 Genres by Popularity")

# 5. Hiển thị
plt.show()
#Insight: Dance, Pop và Rap là 3 thể loại dẫn đầu về độ phổ biến (đạt xấp xỉ 100 điểm).
# Điều này cho thấy xu hướng âm nhạc đại chúng hiện nay tập trung vào các dòng nhạc sôi động, dễ nhảy và có tính bắt tai cao.
```
